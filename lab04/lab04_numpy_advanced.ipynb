### Ćwiczenie 1: Broadcasting — zastosowania biznesowe (20 min)


```python
# Ceny 4 produktów w 3 sklepach
ceny = np.array([[29.99, 31.99, 28.99], # Produkt A
[49.99, 52.99, 47.99], # Produkt B
[14.99, 15.99, 13.99], # Produkt C
[99.99, 104.99, 94.99]]) # Produkt D
sklepy = ['Centrum', 'Galeria', 'Online']
produkty = ['A', 'B', 'C', 'D']
print(f"Kształt macierzy: {ceny.shape}") # (4, 3)
```

```python
# Zadanie 1a
nowe_ceny = ceny * 1.05
print(f"Ceny po inflacji:\n {nowe_ceny}")
```

```python
# Zadanie 1b
rabaty = np.array([0.10, 0.05, 0.15])
ceny_po_rabacie = ceny * (1 - rabaty)
print(f"Ceny po zastosowaniu rabatów dla poszczególnych sklepów:\n {ceny_po_rabacie}")
```

```python
# Zadanie 1c
marza = np.array([
    [0.30], # Produkt A: 30%
    [0.40], # Produkt B: 40%
    [0.25], # Produkt C: 25%
    [0.50]  # Produkt D: 50%
])
koszt_zakupu = ceny / (1 + marza)
print(f"Koszt zakupu po odjeciu marzy dla odpowiednich produktow:\n {koszt_zakupu}")
```

```python
# Zadanie 1d
indeksy_najtanszych = ceny.argmin(axis=1)
for i, indeks in enumerate(indeksy_najtanszych):
    print(f"Produkt {produkty[i]}: {sklepy[indeks]}")
```

### Ćwiczenie 2: Reshape i łączenie danych (20 min)

```python
# Zadanie 2a
# Dane miesięczne z jednego roku (flat)
dane_flat = np.array([120, 135, 98, 142, 167, 189, 201, 178, 156, 143, 198, 221])
# Przekształć na kwartały (4 kwartały × 3 miesiące)
kwartaly = dane_flat.reshape(4, 3)
print(f"Kwartały:\n{kwartaly}")
print(f"Suma per kwartał: {kwartaly.sum(axis=1)}")
print(f"Najlepszy kwartał: Q{kwartaly.sum(axis=1).argmax() + 1}")
```

```python
# Zadanie 2b
# Dane z 3 oddziałów (osobne tablice)
warszawa = np.array([150, 160, 170, 180])
krakow = np.array([120, 130, 140, 150])
wroclaw = np.array([90, 100, 110, 120])
# Połącz w macierz (3 oddziały × 4 kwartały)
wszystkie = np.vstack([warszawa, krakow, wroclaw])
print(f"Wszystkie:\n{wszystkie}")
print(f"Suma per oddział: {wszystkie.sum(axis=1)}")
print(f"Suma per kwartał: {wszystkie.sum(axis=0)}")
```

```python
# Zadanie 2c
sprzedaz_flat = np.array([100, 200, 150, 300, 250, 180, 120, 220, 170, 310, 260, 190])
sprzedaz_macierz = sprzedaz_flat.reshape(6, 2)
srednia_sprzedaz = sprzedaz_macierz.mean(axis=1)
zmiana = sprzedaz_macierz[:, 1] - sprzedaz_macierz[:, 0]
print(f"Macierz sprzedaży:\n{sprzedaz_macierz}\nŚrednia sprzedaz dla każdego produktu:{srednia_sprzedaz}\nZmiana sprzedazy:{zmiana}")
```

### Ćwiczenie 3: Analiza danych finansowych (30 min)

```python
np.random.seed(2026)
# 8 pracowników, dane roczne
pracownicy = ['Anna', 'Jan', 'Ewa', 'Marek', 'Kasia', 'Piotr', 'Zofia', 'Tomek']
pensje = np.array([5500, 7200, 4800, 9100, 6300, 4200, 8500, 5800])
staz = np.array([3, 8, 2, 15, 5, 1, 12, 4])
oceny = np.array([4.2, 3.8, 4.5, 4.0, 4.7, 3.5, 4.3, 3.9])
# Sprzedaż miesięczna 4 produktów przez 6 miesięcy
produkty_nazwy = ['Laptop', 'Tablet', 'Smartfon', 'Akcesorium']
sprzedaz = np.random.randint(20, 200, size=(4, 6))
ceny_produktow = np.array([3500, 1800, 2500, 150])
```

```python
# Zadanie 3a
srednia = pensje.mean()
mediana = np.median(pensje)
odchylenie = pensje.std()
print(f"Średnia:{srednia:.2f}zł\nMediana:{mediana:.2f}zł\nOdchylenie standardowe:{odchylenie:.2f}zł")
nazwy_np = np.array(pracownicy)
powyzej_sredniej = nazwy_np[pensje > srednia]
ponizej_mediany = nazwy_np[pensje < mediana]
print(f"Osoby zarabiające powyżej średniej:{powyzej_sredniej}\nOsoby zarabiające poniżej mediany:{ponizej_mediany}")
rozstep = pensje.max() - pensje.min()
q3, q1 = np.percentile(pensje, [75, 25])
iqr = q3 - q1
print(f"Rozstep:{rozstep}zł\nRozstep międzykwartylowy:{iqr}zł")
```

```python
# Zadanie 3b
import numpy as np
premie = np.where(oceny >= 4.5, 2000, 
                  np.where(oceny >= 4.0, 1000, 0))
nowe_wynagrodzenie = pensje + premie
indeks_max_premii = premie.argmax()
osoba_z_max_premia = pracownicy[indeks_max_premii]
print("Premie dla pracowników:")
for i in range(len(pracownicy)):
    print(f"{pracownicy[i]}: {premie[i]} zł (Ocena: {oceny[i]})")
print(f"Najwyższą premię otrzymał(a): {osoba_z_max_premia} ({premie[indeks_max_premii]} zł)")
print(f"Średnie nowe wynagrodzenie: {nowe_wynagrodzenie.mean():.2f} zł")
```

```python
# Zadanie 3c
import numpy as np
indeksy_rosnaco = np.argsort(pensje)
kolejnosc = indeksy_rosnaco[::-1]
print(f"{'Poz.':<5} | {'Imię':<8} | {'Pensja':<10}| {'Staż':<5}")
for i, idx in enumerate(kolejnosc, 1):
    print(f"{i:<5} | {pracownicy[idx]:<8} | {pensje[idx]:>6} zł | {staz[idx]:>2} lat")
```

```python
# Zadanie 3d
import numpy as np
kor_staz_pensja = np.corrcoef(staz, pensje)[0, 1]
kor_staz_ocena = np.corrcoef(staz, oceny)[0, 1]
sila_pensja = abs(kor_staz_pensja)
sila_ocena = abs(kor_staz_ocena)
print(f"Korelacja Straż vs Pensja: {kor_staz_pensja:.4f}\nKorelacja Straż vs Ocena: {kor_staz_ocena:.4f}")
print("\nKtóry związek jest silniejszy:")
if sila_pensja > sila_ocena:
    print("Związek między stażem a pensją.")
else:
    print("Związek między stażem a oceną pracowniczą.")
```

```python
# Zadanie 3e
import numpy as np
ceny_kol = ceny_produktow.reshape(4, 1)
przychod = sprzedaz * ceny_kol
suma_per_produkt = przychod.sum(axis=1)
najlepszy_prod_idx = suma_per_produkt.argmax()
suma_per_miesiac = przychod.sum(axis=0)
najlepszy_msc_idx = suma_per_miesiac.argmax()
przychod_I_polowa = przychod[:, :3].sum()
przychod_II_polowa = przychod[:, 3:].sum()
print("Macierz przychodów (4x6):\n", przychod)
print(f"Najlepszy produkt: {produkty_nazwy[najlepszy_prod_idx]} (Przychód: {suma_per_produkt[najlepszy_prod_idx]} zł)")
print(f"Najlepszy miesiąc (indeks): {najlepszy_msc_idx} (Suma: {suma_per_miesiac[najlepszy_msc_idx]} zł)")
print(f"Suma I kwartał (m-ce 0-2): {przychod_I_polowa} zł")
print(f"Suma II kwartał (m-ce 3-5): {przychod_II_polowa} zł")
print(f"Trend: {'Wzrostowy' if przychod_II_polowa > przychod_I_polowa else 'Spadkowy'}")
```

```python
import numpy as np
np.random.seed(42)
wiek = np.clip(np.random.normal(35, 10, 100).astype(int), 18, None)
wydatki = np.clip(np.random.normal(500, 200, 100), 0, None)
wizyty = np.random.poisson(5, 100)
print(f"{'Klient':<8} | {'Wiek':<5} | {'Wydatki':<10} | {'Wizyty':<5}")
for i in range(5):
    print(f"{i+1:<8} | {wiek[i]:<5} | {wydatki[i]:>7.2f} zł | {wizyty[i]:>5}")
print("\nPodstawowe statystyki wygenerowanej grupy:")
print(f"Średni wiek: {wiek.mean():.1f}")
print(f"Średnie wydatki: {wydatki.mean():.2f} zł")
print(f"Max wizyt: {wizyty.max()}")
```

```python
import numpy as np
print("--- STATYSTYKI OPISOWE ---")
print(f"Wiek     - Średnia: {wiek.mean():.1f}, Mediana: {np.median(wiek):.1f}, Std: {wiek.std():.1f}")
print(f"Wydatki  - Średnia: {wydatki.mean():.2f}, Mediana: {np.median(wydatki):.2f}, Std: {wydatki.std():.2f}")
print(f"Wizyty   - Średnia: {wizyty.mean():.1f}, Max: {wizyty.max()}, Min: {wizyty.min()}")
kor_wiek_wydatki = np.corrcoef(wiek, wydatki)[0, 1]
kor_wizyty_wydatki = np.corrcoef(wizyty, wydatki)[0, 1]
print("\n--- ANALIZA ZWIĄZKÓW (KORELACJA) ---")
print(f"Wiek vs Wydatki: {kor_wiek_wydatki:.3f}")
print(f"Wizyty vs Wydatki: {kor_wizyty_wydatki:.3f}")
sr = wydatki.mean()
sd = wydatki.std()
segmenty = np.where(wydatki > sr + sd, 'WYSOKI', 
                   np.where(wydatki < sr - sd, 'NISKI', 'ŚREDNI'))
nazwy_segmentow, liczebnosc = np.unique(segmenty, return_counts=True)
print("\n--- WYNIKI SEGMENTACJI ---")
for nazwa, ile in zip(nazwy_segmentow, liczebnosc):
    print(f"Segment {nazwa}: {ile} klientów")
print(f"\nŚredni wiek klientów o wysokich wydatkach: {wiek[segmenty == 'WYSOKI'].mean():.1f} lat")
```
