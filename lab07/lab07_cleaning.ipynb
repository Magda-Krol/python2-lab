## Ćwiczenie 3: Pełny pipeline czyszczenia (25 min)

```python
df_clean = pd.DataFrame(data)
df_clean['wynagrodzenie'] = df_clean['wynagrodzenie'].replace('brak', np.nan)
df_clean['wynagrodzenie'] = pd.to_numeric(df_clean['wynagrodzenie'], errors='coerce')

# KROK 2: Usuwanie duplikowanych wierszy
df_clean = df_clean.drop_duplicates().reset_index(drop=True)

# KROK 3: Uzupełnianie braków w wynagrodzeniu medianą
mediana_wyn = df_clean['wynagrodzenie'].median()
df_clean['wynagrodzenie'] = df_clean['wynagrodzenie'].fillna(mediana_wyn)

# KROK 4: Uzupełnianie braków w ocenie rocznej średnią
srednia_ocena = round(df_clean['ocena_roczna'].mean(), 2)
df_clean['ocena_roczna'] = df_clean['ocena_roczna'].fillna(srednia_ocena)

# KROK 5: Standaryzacja imion
df_clean['imie'] = df_clean['imie'].str.strip().str.title()

# KROK 6: Standaryzacja nazw działów
df_clean['dzial'] = df_clean['dzial'].str.strip().str.title()
df_clean['dzial'] = df_clean['dzial'].replace('Hr', 'HR', regex=False)
df_clean['dzial'] = df_clean['dzial'].replace('It', 'IT', regex=False)

# KROK 7: Konwersja daty zatrudnienia
df_clean['data_zatrudnienia'] = pd.to_datetime(df_clean['data_zatrudnienia'], errors='coerce')

# === WERYFIKACJA ===
print("=== WERYFIKACJA ===")
print(f"Shape: {df_clean.shape}")
print(f"\nDtypes:\n{df_clean.dtypes}")
print(f"\nNaN per kolumna:\n{df_clean.isna().sum()}")
print(f"\nUnikalne działy: {sorted(df_clean['dzial'].unique())}")
print(f"\nPierwsze imiona: {df_clean['imie'].tolist()[:8]}")
```

```python
# Pytanie 1: Jakie jest średnie wynagrodzenie per dział?
srednie_wynagrodzenie = df_clean.groupby('dzial')['wynagrodzenie'].mean().round(2)
print("Średnie wynagrodzenie per dział:")
print(srednie_wynagrodzenie)
```

```python
# Pytanie 2: Jaka jest średnia ocena per dział?
srednia_ocena_dzial = df_clean.groupby('dzial')['ocena_roczna'].mean().round(2)
print("\nŚrednia ocena per dział:")
print(srednia_ocena_dzial)
```

```python
# Pytanie 3: Kto jest najlepiej opłacany (top 3)?
top_3_place = df_clean.nlargest(3, 'wynagrodzenie')[['imie', 'dzial', 'wynagrodzenie']]
print("\nTop 3 najlepiej opłacanych pracowników:")
print(top_3_place)
```

```python
# Pytanie 4: Ilu pracowników jest w każdym dziale?
liczebnosc_dzialow = df_clean['dzial'].value_counts()
print("\nLiczba pracowników w każdym dziale:")
print(liczebnosc_dzialow)
```

```python
# 1. Ile lat pracuje najdłuższy staż?
dzis = pd.to_datetime('2026-04-20')
najwczesniejsza_data = df_clean['data_zatrudnienia'].min()

staz_dni = (dzis - najwczesniejsza_data).days
staz_lata = round(staz_dni / 365.25, 1)

print(f"Najdłuższy staż w firmie wynosi: {staz_lata} lat")
print(f"Najstarsze zatrudnienie z dnia: {najwczesniejsza_data.date()}")
```

```python
# 2. Kto ma najwyższą ocenę w każdym dziale? (groupby + idxmax)
idx_najlepsi = df_clean.groupby('dzial')['ocena_roczna'].idxmax()
najlepsi_pracownicy = df_clean.loc[idx_najlepsi, ['dzial', 'imie', 'ocena_roczna']]
```

```python
# 3. Czy jest korelacja między stażem a wynagrodzeniem?
df_clean['staz_dni'] = (dzis - df_clean['data_zatrudnienia']).dt.days


korelacja = df_clean['staz_dni'].corr(df_clean['wynagrodzenie']).round(2)

print(f"\nKorelacja między stażem a wynagrodzeniem: {korelacja}")

if korelacja > 0.5:
    print("Interpretacja: Istnieje silna dodatnia korelacja (staż wpływa na płacę).")
elif korelacja < -0.5:
    print("Interpretacja: Istnieje silna ujemna korelacja.")
else:
    print("Interpretacja: Korelacja jest słaba – staż pracy nie determinuje wysokości zarobków.")
```
