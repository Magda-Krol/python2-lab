```python
import pandas as pd
import numpy as np
# Series z listy — domyślny indeks (0, 1, 2, ...)
oceny = pd.Series([4.5, 3.0, 5.0, 3.5, 4.0])
print(oceny)
```

```python
#Zadanie 1a
pensje = pd.Series({
    'Anna':  5500,
    'Jan':    7200,
    'Ewa':   4800,
    'Marek':    9100,
    'Kasia':     6300
})
```

```python
#Zadanie 1b
#1.
osoba_max = pensje.idxmax()
print(f"Osoba zarabiająca najwięcej: {osoba_max}")
#2.
srednia_pensja = pensje.mean()
print(f"Średnia pensja: {srednia_pensja}")
#3.
maska = pensje > srednia_pensja
osoby_powyzej_sredniej = pensje[maska].index
print(f"Osoby zarabiające powyżej średniej: {list(osoby_powyzej_sredniej)}")
#4.
maska = pensje < 6000
osoby_ponizej_sredniej = pensje[maska].index
print(f"Ilość osób zarabiających poniżej średniej: {len(osoby_ponizej_sredniej)}")
```

```python
#Zadanie 1c
#1.
print(f"Pensja po podwyżce 10%: ")
po_podwyzce_10_procent = pensje * 1.10
print(po_podwyzce_10_procent.to_string())
#2.
print(f"\nPensja po premii 500zł: ")
po_premii_500 = pensje + 500
print(po_premii_500.to_string())
#3.
print(f"\nNowa pensja: ")
nowa_pensja = (pensje * 1.10) + 500
print(nowa_pensja.to_string())
#4.
print(f"\nRanking nowych pensji od najwyższej: ")
print(nowa_pensja.sort_values(ascending=False).to_string())
```

#### Ćwiczenie 4: Pytania biznesowe + commit (15 min)

```python
tips = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv')
```

```python
tips['tip_pct'] = (tips['tip'] / tips['total_bill'] * 100).round(1)
```

```python
# 1. Oblicz średni rachunek (total_bill)
sredni_rachunek = tips['total_bill'].mean()
print(f"1. Średni rachunek: {sredni_rachunek:.2f}")
```

```python
# 2. Oblicz średni napiwek w procentach (tip_pct)
sredni_napiwek_pct = tips['tip_pct'].mean()
print(f"2. Średni napiwek (%): {sredni_napiwek_pct:.2f}%")
```

```python
# 3. Wyświetl który dzień ma najwięcej rachunków (value_counts na 'day')
dni_counts = tips['day'].value_counts()
print(f"\nLiczba rachunków w podziale na dni:\n {dni_counts}")
print(f"Najwięcej rachunków jest w: {dni_counts.idxmax()}")
```

```python
# 4. Policz ile rachunków jest od palaczy (value_counts na 'smoker')
palacze_counts = tips['smoker'].value_counts()
print(f"\nLiczba rachunków (palacze vs niepalący): {palacze_counts}")
```

```python
# 5. Wyświetl describe() tylko dla kolumny 'tip' — jaka jest mediana napiwku?
tip_stats = tips['tip'].describe()
mediana_napiwku = tips['tip'].median()
print(f"\nStatystyki dla kolumny 'tip': {tip_stats}")
print(f"Mediana napiwku wynosi: {mediana_napiwku}")
```

```python
# 6. Jaki procent rachunków pochodzi z obiadu (Dinner)?
procent_czasu = tips['time'].value_counts(normalize=True) * 100
print(f"\nProcentowy udział porposiłków: {procent_czasu}")
print(f"Z obiadu (Dinner) pochodzi: {procent_czasu['Dinner']:.2f}% rachunków")
```

```python
# 7. Ile unikalnych rozmiarów grup jest w danych? (nunique na 'size')
unikalne_rozmiary = tips['size'].nunique()
print(f"\nLiczba unikalnych rozmiarów grup: {unikalne_rozmiary}")
```
