#### Ćwiczenie 3: Operacje wektorowe — analiza biznesowa (25min)

```python
# Produkty w sklepie internetowym
nazwy = ['Laptop', 'Tablet', 'Smartfon', 'Słuchawki', 'Ładowarka',
'Klawiatura', 'Mysz', 'Monitor', 'Kamera', 'Drukarka']
ceny = np.array([3499, 1899, 2499, 349, 89, 249, 129, 1499, 599, 899])
ilosci_magazyn = np.array([15, 42, 78, 156, 312, 89, 201, 23, 45, 34])
oceny_klientow = np.array([4.7, 4.2, 4.5, 4.8, 3.9, 4.1, 4.3, 4.6, 3.8, 3.5])
```

```python
print(ceny.mean())
```

```python
print(ceny.max())
```

```python
print(ceny.min())
```

```python
nazwy = np.array(nazwy)
print(f"najdrozszy produkt: {nazwy[ceny.argmax()]}")
```

```python
nazwy = np.array(nazwy)
print(f"najtanszy produkt: {nazwy[ceny.argmin()]}")
```

```python
wartosc_produktow = ceny * ilosci_magazyn 
for nazwa, wartosc_produktu in zip(nazwy, wartosc_produktow):
    print(f"{nazwa} łacznie: {wartosc_produktu}zł")
print(f"wartosc magazynu: {np.sum(ceny * ilosci_magazyn)}zł")
```

```python
nazwy = np.array(nazwy)
promocja_swiateczna = ceny * 0.85
for nazwa, promocja in zip(nazwy, promocja_swiateczna):
    print(f"{nazwa}: {promocja:.2f}zł")
```

```python
nazwy = np.array(nazwy)
po_vat = ceny * 1.23
for nazwa, vat in zip(nazwy, po_vat):
    print(f"{nazwa}: {vat:.2f}zł")
```

```python
print(f"ilosc powyzej 1000: {(ceny > 1000).sum()}")
```

```python
maska = oceny_klientow >= 4.5
nazwy_wybrane = nazwy[maska]
ceny_wybrane = ceny[maska]
for nazwa, cena in zip(nazwy_wybrane, ceny_wybrane):
    print(f"{nazwa}: {cena}zł")
```

```python
maska = ceny < 200
nazwy_wybrane = nazwy[maska]
ceny_wybrane = ceny[maska]
ilosci_wybrane = ilosci_magazyn[maska]
for nazwa, cena, ilosc in zip(nazwy_wybrane, ceny_wybrane, ilosci_wybrane):
    print(f"{nazwa}: {ilosc}szt, {cena}zł")
```

```python
maska = ilosci_magazyn < 30
nazwy_wybrane = nazwy[maska]
ilosci_wybrane = ilosci_magazyn[maska]
for nazwa, ilosc in zip(nazwy_wybrane, ilosci_wybrane):
    print(f"{nazwa}: {ilosc}szt.")
```

```python
ceny_niski_stan = ceny[maska]
print(ceny_niski_stan)
print(f"Łączna wartość tych produktów: {(ilosci_wybrane * ceny_niski_stan).sum()}")
```

```python
caly_magazyn = (ceny * ilosci_magazyn).sum()
procent_magazynu = ((ilosci_wybrane * ceny_niski_stan).sum()/caly_magazyn)*100
print(f"Ile % watrosci całego magazynu: {procent_magazynu: .2f}%")
```

```python
nakaz_1 = oceny_klientow >= 4.5
nakaz_2 = ceny < 1000
wynik = nazwy[nakaz_1 & nakaz_2]
print(wynik)
```

```python
sr_wieksza = oceny_klientow >= 4.0
sr_mniejsza = oceny_klientow < 4.0
wynik_mniejsza = ceny[sr_mniejsza].mean()
wynik_wieksza = ceny[sr_wieksza].mean()
print(f"Średnia cena przy ocenie >=4.0 vs <4.0: {wynik_wieksza:.2f}zł , {wynik_mniejsza:.2f}zł")
```

```python
# Przygotowanie danych
import time
rozmiar = 1_000_000
lista = list(range(rozmiar))
tablica = np.array(lista)
```

```python
# Test 1: Pomnóż przez 2
start = time.perf_counter()
wynik_lista = [x * 2 for x in lista]
czas_lista = time.perf_counter() - start
start = time.perf_counter()
wynik_numpy = tablica * 2
czas_numpy = time.perf_counter() - start
print(f"=== Mnożenie ×2 ({rozmiar:,} elementów) ===")
print(f"Lista: {czas_lista*1000:.1f} ms")
print(f"NumPy: {czas_numpy*1000:.1f} ms")
print(f"NumPy jest {czas_lista/czas_numpy:.0f}× szybszy!")
```

```python
# Test 2: Suma
start = time.perf_counter()
suma_lista = sum(lista)
czas_lista = time.perf_counter() - start
start = time.perf_counter()
suma_numpy = tablica.sum()
czas_numpy = time.perf_counter() - start
print(f"\n=== Suma ({rozmiar:,} elementów) ===")
print(f"Lista: {czas_lista*1000:.1f} ms")
print(f"NumPy: {czas_numpy*1000:.1f} ms")
print(f"NumPy jest {czas_lista/czas_numpy:.0f}× szybszy!")
```

```python
# Test 3: Filtrowanie (elementy > 500000)
start = time.perf_counter()
filtr_lista = [x for x in lista if x > 500000]
czas_lista = time.perf_counter() - start
start = time.perf_counter()
filtr_numpy = tablica[tablica > 500000]
czas_numpy = time.perf_counter() - start
print(f"\n=== Filtrowanie > 500k ({rozmiar:,} elementów) ===")
print(f"Lista: {czas_lista*1000:.1f} ms")
print(f"NumPy: {czas_numpy*1000:.1f} ms")
print(f"NumPy jest {czas_lista/czas_numpy:.0f}× szybszy!")
```
